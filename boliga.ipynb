```python
def clean_boligtype_adresse(bolig_type_adresse_series):
    cleaned_boligtyper = []
    cleaned_adresser = []
    
    for bolig_type in bolig_type_adresse_series:
        bolig = bolig_type.split(' ')
        boligtype = bolig[0]
        fixed_bolig_type = fix_duplicate_inside(boligtype)
        new_bolig = ' '.join([fixed_bolig_type]) 
        adresse = ' '.join(bolig[1:])
        
        cleaned_boligtyper.append(new_bolig)
        cleaned_adresser.append(adresse)
    
    return pd.Series(cleaned_boligtyper), pd.Series(cleaned_adresser)

cleaned = clean_boligtype_adresse(bolig_type_adresse)
cleaned_boligtyper, cleaned_adresser = cleaned
```
