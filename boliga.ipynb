```python
# function to clean duplicated string values in 'Boligtype & Adresse' column and split into two new columns
def clean_boligtype_adresse(bolig_type_adresse_series):
    cleaned_boligtyper = []
    cleaned_adresser = []
        
    for bolig_type in bolig_type_adresse_series:
        bolig_type_list = bolig_type.split(' ')
        bolig_type_list_first_element = bolig_type_list[0]
        # imported function to fix duplicates inside the first element 
        fixed_bolig_type = fix_duplicate_inside(bolig_type_list_first_element)
        fixed_adresse = ' '.join(bolig_type_list[1:])
        cleaned_boligtyper.append(fixed_bolig_type)
        cleaned_adresser.append(fixed_adresse)
    
    return pd.Series(cleaned_boligtyper), pd.Series(cleaned_adresser)

cleaned_boligtyper, cleaned_adresser = clean_boligtype_adresse(df_copy['Boligtype & Adresse'])
df_copy['bolig_type'] = cleaned_boligtyper.values
df_copy['adresse'] = cleaned_adresser.values
```

```python
def extract_dato_type(dato_og_type_series):
    """Extract date and type from 'Dato & Type' column.
    
    Input format: "30-11-2025Alm. Salg" or "30-11-2025Alm.Salg"
    Output: two Series - dates (as datetime.date) and types (as strings)
    """
    dates = []
    types = []
    
    for entry in dato_og_type_series.fillna(''):
        # Find the date pattern (DD-MM-YYYY)
        match = re.search(r'\d{2}-\d{2}-\d{4}', str(entry))
        
        if match:
            date_str = match.group()
            # Convert to date object
            date_obj = datetime.strptime(date_str, '%d-%m-%Y').date()
            dates.append(date_obj)
            
            # Extract the type (everything after the date)
            type_str = entry[match.end():].strip()
            types.append(type_str)
        else:
            # If no date found, append NaN/empty
            dates.append(None)
            types.append('')
    
    return pd.Series(dates, index=dato_og_type_series.index), pd.Series(types, index=dato_og_type_series.index)

# Apply the extraction function
dato_series, type_series = extract_dato_type(df_copy['Dato & Type'])
df_copy['Dato'] = dato_series
df_copy['Type'] = type_series

# Verify the results
print("Sample extracted data:")
print(df_copy[['Dato & Type', 'Dato', 'Type']].head(10))
```
